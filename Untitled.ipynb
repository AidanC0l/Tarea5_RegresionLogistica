```python
#Cargamos las librerias necesarias
import pandas as pd #Manejo y analisis de estruturas de datos
import numpy as np #Calculo numero y el analisis de datos
import seaborn as sns #Creacion de graficos estadisticos
import matplotlib.pyplot as plt #creacion de graficos en dos dimensiones

#Librerias para implementar modelos de ciencia de datos
from sklearn import metrics
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
from sklearn.feature_selection import SelectKBest, f_classif
from sklearn.model_selection import GridSearchCV
```

```python
#Lectura de datos en Python y lo cargamos en la variable Datos
Datos = pd.read_csv('Titanic-Dataset.csv')
```

```python
#Realizamos la lectura de los datos para el analisis exploratorio
Datos.head(40)
```

```python
# Resumen general del dataset
print(Datos.info())
print(Datos.describe())
```

```python
# Visualización inicial: supervivencia por género
sns.countplot(data=Datos, x='Survived', hue='Sex')
plt.title('Supervivencia por género')
plt.show()
```

# Procesamiento de datos.

```python
# Identificar valores nulos
print(Datos.isnull().sum())
```

```python
# Tratamiento de valores faltantes
Datos['Age'].fillna(Datos['Age'].median())
Datos['Embarked'].fillna(Datos['Embarked'].mode())
```

```python
# Eliminación de columnas irrelevantes
Datos.drop(['Cabin', 'Ticket', 'Name'], axis=1, inplace=True)
```

```python
# Convertimos las variables categóricas a numéricas
Datos = pd.get_dummies(Datos, columns=['Sex', 'Embarked'], drop_first=True)
```

```python
# Verificamos de nuevo los valores nulos
print(Datos.isnull().sum())
```

# Selección de características.

```python
# Separamos las variables independientes y dependientes
X = Datos.drop('Survived', axis=1)
y = Datos['Survived']
```

```python
# Seleccionamos las características más relevantes
selector = SelectKBest(score_func=f_classif, k=5)
X_new = selector.fit_transform(X, y)
print('Puntajes de características:', selector.scores_)
```

```python
# Dividimos el DataSet
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_new, y, test_size=0.2, random_state=42)
print(f'Tamaño del conjunto de entrenamiento: {X_train.shape}')
print(f'Tamaño del conjunto de prueba: {X_test.shape}')
```

```python
# Definimos los hiperparámetros a probar
param_grid = {'C': [0.1, 1, 10], 'solver': ['liblinear', 'lbfgs']}
logreg = LogisticRegression()
```

```python
# Buscamos los mejores hiperparámetros
grid_search = GridSearchCV(logreg, param_grid, cv=5, scoring='accuracy')
grid_search.fit(X_train, y_train)
```

```python
# Pedimos los mejores parametros para un mejor modelo
best_model = grid_search.best_estimator_
print('Mejores parámetros:', grid_search.best_params_)
```

# Evaluación del modelo.

```python
# Realizamos las predicciones
y_pred = best_model.predict(X_test)
```

```python
# Métricas
print('Matriz de confusión:\n', confusion_matrix(y_test, y_pred))
print('Reporte de clasificación:\n', classification_report(y_test, y_pred))
print('Precisión:', accuracy_score(y_test, y_pred))
```

# Graficas.

```python
# Matriz de confusión
sns.heatmap(confusion_matrix(y_test, y_pred), annot=True, fmt='d', cmap='Reds')
plt.title('Matriz de Confusión')
plt.xlabel('Predicción')
plt.ylabel('Realidad')
plt.show()
```

```python
# Curva de precisión del modelo
from sklearn.metrics import roc_curve, roc_auc_score

y_pred_proba = best_model.predict_proba(X_test)[:, 1]
fpr, tpr, _ = roc_curve(y_test, y_pred_proba)
roc_auc = roc_auc_score(y_test, y_pred_proba)

plt.plot(fpr, tpr, label=f'AUC = {roc_auc:.2f}')
plt.plot([0, 1], [0, 1], 'k--')
plt.title('Curva ROC')
plt.xlabel('FPR')
plt.ylabel('TPR')
plt.legend()
plt.show()
```
